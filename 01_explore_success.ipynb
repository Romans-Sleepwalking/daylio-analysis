# Explore sucess

```python
import pandas as pd
import seaborn as sns
import numpy as np
import shap
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
import warnings
warnings.filterwarnings("ignore")
# Specify the dataset by date
DAYLIO_DATE = '11_20'
```

### Read 

```python
# Read CSV file exported from the Daylio app
filename = f'transformed_data_{DAYLIO_DATE}.csv'
df = pd.read_csv(filename)
df.set_index('datetime', inplace=True)
df.info()
```

### Dynamics of Success

```python
plt.figure(figsize=(10, 6), dpi=400)
plt.plot(df.index, df['success'], label='Success Dynamics', color='blue')
plt.title('Success Scores Daily')
plt.xlabel('Time in Days')
plt.ylabel('Success Scores')
plt.xticks([])
plt.legend()
plt.show()
```

### Calculate average effect on success

```python
# Select only the numeric columns (excluding 'datetime')
numeric_columns = df.iloc[:, 1:].select_dtypes(include=['float64', 'int64'])
# Calculate the correlation matrix
correlation_matrix = numeric_columns.corrwith(df['success'])
# Create a DataFrame with correlations
correlation_df = correlation_matrix.reset_index()
correlation_df.columns = ['Feature', 'Correlation']
correlation_df.sort_values(by='Correlation', ascending=False, inplace=True)
# Create a bar plot
plt.figure(figsize=(10, 8))
sns.barplot(data=correlation_df, x='Correlation', y='Feature', palette="viridis")
# Set the height of the bars
bar_height = 0.5  # Adjust as needed
# Customize the plot
plt.xlabel("Correlation")
plt.ylabel("Feature")
plt.title("Correlations between 'success' and other features")
# Adjust the y-axis labels and set the bar height
plt.yticks(range(len(correlation_df['Feature'])), correlation_df['Feature'], fontsize=10)
# Show the plot
plt.show()
```

### Regression Feature Importance

```python
X = df.iloc[:, 2:].copy()
y = df.iloc[:, 0].copy()
X.fillna(X.mean(), inplace=True)
# Assuming 'X' is your feature matrix and 'y' is your target variable
model = RandomForestRegressor()
model.fit(X, y)
# Get feature importances
feature_importances = model.feature_importances_
# Calculate the correlation between each feature and the target variable
correlation_with_target = X.apply(lambda feature: feature.corr(y))
# Multiply feature importances by the sign of the correlation
signed_feature_importances = feature_importances * correlation_with_target.apply(lambda x: 1 if x >= 0 else -1)
# Create a DataFrame to store feature names and their signed importance scores
feature_importance_df = pd.DataFrame({'Feature': X.columns, 'Importance': signed_feature_importances})
# Sort the DataFrame by signed importance score in descending order
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)
# Set the size of the plot
plt.figure(figsize=(12, 8))
# Plot the feature importance with signed values
plt.barh(feature_importance_df['Feature'], feature_importance_df['Importance'])
plt.xlabel('Importance Score (Signed)')
plt.title('Feature Importance (Signed)')
# Show the plot
plt.show()
```

### XGBoost Predictability

```python
X = df.iloc[:, 1:].copy()
y = df.iloc[:, 0].copy()
X.fillna(X.mean(), inplace=True)
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Creating an XGBoost model
model = xgb.XGBRegressor(n_estimators=80, max_depth=4, eta=0.02)
model.fit(X_train, y_train)
# Assuming 'model' is your trained XGBoost regressor
y_pred = model.predict(X_test)  # Predicting the continuous values for the test set
# Calculate mean squared error
mse = mean_squared_error(y_test, y_pred)
# Display the mean squared error
print(f"Mean Squared Error: {mse}")
```

```python
indices = np.arange(len(y_test))
bar_width = 0.35
plt.figure(figsize=(12, 6))
plt.bar(indices - bar_width/2, y_test, bar_width, label='My Evaluation of Success', color='blue')
plt.bar(indices + bar_width/2, y_pred, bar_width, label='Model Prediction', color='orange')
plt.xlabel('Data Points')
plt.ylabel('Values')
plt.title('Comparison of True and Predicted Values')
plt.legend()
plt.show()
```

```python
# Plotting one of the decision trees
xgb.plot_tree(model, num_trees=2)
fig = plt.gcf()
fig.set_size_inches(150, 100)
fig.savefig('tree.png')
```

```python
# Explain the model's predictions using SHAP values
explainer = shap.Explainer(model)
shap_values = explainer.shap_values(X_test)
shap.summary_plot(shap_values, X_test, plot_type="bar", show=False)
plt.show()
```
